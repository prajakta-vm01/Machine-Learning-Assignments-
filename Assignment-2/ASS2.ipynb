User Defined Loop-Based Statistics

```python
import math

def calculate_statistics(column_data):

    n = len(column_data)

    # Mean
    total = 0
    for value in column_data:
        total += value
    mean = total / n

    # Median
    sorted_data = sorted(column_data)
    if n % 2 == 0:
        median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        median = sorted_data[n//2]

    # Mode
    freq = {}
    for value in column_data:
        freq[value] = freq.get(value, 0) + 1
    mode = max(freq, key=freq.get)

    # Variance
    variance_sum = 0
    for value in column_data:
        variance_sum += (value - mean) ** 2
    variance = variance_sum / n

    # Standard Deviation
    std_dev = math.sqrt(variance)

    # Range
    range_value = max(column_data) - min(column_data)

    return mean, median, mode, variance, std_dev, range_value
```

```python
for column in df.columns:
    print("\nFeature:", column)

    data = list(df[column])

    mean, median, mode, variance, std_dev, range_value = calculate_statistics(data)

    print("Mean:", mean)
    print("Median:", median)
    print("Mode:", mode)
    print("Variance:", variance)
    print("Standard Deviation:", std_dev)
    print("Range:", range_value)
```

User Defined Statistics Class

```python
class Statistics:

    def __init__(self, data):
        self.data = sorted(data)
        self.n = len(data)

    def mean(self):
        return sum(self.data) / self.n

    def median(self):
        if self.n % 2 == 0:
            return (self.data[self.n//2 - 1] + self.data[self.n//2]) / 2
        else:
            return self.data[self.n//2]

    def variance(self):
        mean = self.mean()
        return sum((x - mean)**2 for x in self.data) / self.n

    def std_dev(self):
        return math.sqrt(self.variance())

    def data_range(self):
        return max(self.data) - min(self.data)
```

Use Statistics Class

```python
for column in df.columns:
    print("\nUsing Class for:", column)

    stats = Statistics(list(df[column]))

    print("Mean:", stats.mean())
    print("Median:", stats.median())
    print("Variance:", stats.variance())
    print("Standard Deviation:", stats.std_dev())
    print("Range:", stats.data_range())
```
