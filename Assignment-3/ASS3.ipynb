Assignment No.3: Visualize the data using Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Creating the Dataset

```python
data={
    'student':['Prajakta','Srushti','Rutika','Rajnandini','Pooja', 'Bhoomi',
               'Purva','Vanshika','Tanisha'],
    'Machine learning':[94,93,90,91,89,92,90,91,88],
    'Data Structure':[92,93,93,90,90,91,92,89,88],
    'Python':[90,91,89,88,92,90,91,89,87],
     'Artificial Intelligence':[93,92,90,91,89,91,90,90,88],
    'DBMS':[91,92,90,70,91,90,92,88,87]
}
```

```python
df=pd.DataFrame(data)
```

```python
print(df)
```

Scatter Plot

This chart shows the relationship between Machine Learning and Data Structure marks.

```python
plt.scatter(df['Machine learning'],df['Data Structure'])
plt.title('ML vs Data Structure - scatter plot')
plt.xlabel('Machine learning')
plt.ylabel('Data Structure')
plt.show()
```

Bar Plot

This chart compares Machine Learning marks of all students.

```python
plt.bar(df['student'], df['Machine learning'])
plt.xlabel('Students')
plt.ylabel('Machine Learning Marks')
plt.title('Bar Plot: ML Marks')
plt.show()
```

Line Chart

This chart shows variation of marks across students for different subjects.

```python
plt.figure()
plt.plot(df['student'], df['Machine learning'], label='ML')
plt.plot(df['student'], df['Data Structure'], label='DS')
plt.plot(df['student'], df['Python'], label='Python')
plt.legend()
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks Trend Across Subjects')
plt.show()
```

Box Plot

This chart shows the distribution of marks including median and outliers.

```python
plt.boxplot([
    df['Machine learning'],
    df['Data Structure'],
    df['Python'],
    df['Artificial Intelligence'],
    df['DBMS']
])
plt.xticks([1,2,3,4,5],['ML','DS','Python','AI', 'DBMS'])
plt.ylabel('Marks')
plt.title('Marks Distribution - Box Plot')
plt.show()
```

Pie Chart

This chart shows percentage contribution of Python marks by each student.

```python
plt.figure()
plt.pie(df['Python'], labels=df['student'], autopct='%1.1f%%')
plt.title('Python Marks Contribution - Pie Chart')
plt.show()
```

Histogram

This chart shows the frequency distribution of Python marks.

```python
plt.figure()
plt.hist(df['Python'])
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Python Marks Histogram')
plt.show()
```

Correlation Heatmap

This chart shows correlation between subjects using color intensity.

```python
corr = df[['Machine learning','Data Structure','Python',
           'Artificial Intelligence','DBMS']].corr()

plt.figure()
plt.matshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.title('Correlation Heatmap', pad=20)
plt.show()
```
